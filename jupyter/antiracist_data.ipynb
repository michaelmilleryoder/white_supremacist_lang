```python
import os
import json
import pandas as pd
from tqdm.notebook import tqdm

dirpath = '../data/antiracist/twitter/'
dfs = []
for fname in tqdm(sorted(os.listdir(dirpath))):
    with open(os.path.join(dirpath, fname)) as f:
        dfs.append(pd.json_normalize([json.loads(line) for line in f.read().splitlines()]))
tweets = pd.concat(dfs).reset_index(drop=True)
tweets['created_at'] = pd.to_datetime(tweets['created_at'])
tweets.info()
tweets
```

```python
# Tokenize, preprocess
import re
from nltk.tokenize import TweetTokenizer
tokenizer = TweetTokenizer(strip_handles=True)

def remove_urls(text, urls):
    new_text = text
    urls = [entity['url'] for entity in urls]
    for url in urls:
        new_text = new_text.replace(url, '<URL>')
    return new_text

def process_text(text, user_mentions, urls):
    new_text = text
    if isinstance(urls, list):
        new_text = remove_urls(new_text, urls)
    new_text = ' '.join(tokenizer.tokenize(new_text)).replace('RT', '').replace('…', '')
    new_text = re.sub(r'^ : ', '', new_text)
    return new_text.lower()

tweets['processed_text'] = [process_text(text, user_mentions, urls) for text, user_mentions, urls in tqdm(zip(
    tweets['text'], tweets['entities.mentions'], tweets['entities.urls']), total=len(tweets))]

tweets[['processed_text', 'text']].sample(20)
```

# Reddit data (forum data)
Scraped by get_reddit.py

```python
# Load data
import os
import re
import pandas as pd
from tqdm.notebook import tqdm

dirpath = '../data/antiracist/reddit_comments'
fpaths = sorted([fname for fname in os.listdir(dirpath)])
dfs = []
for fname in tqdm(fpaths):
    print(fname)
    fpath = os.path.join(dirpath, fname)
    sub = pd.read_json(fpath, orient='table')
    subreddit = fname.split('.')[0]
    dfs.append(sub.assign(subreddit=subreddit.lower()))
    
reddit = pd.concat(dfs).reset_index(drop=True)
reddit['timestamp'] = pd.to_datetime(reddit.created_utc, unit='s', utc=True)
reddit.info()
reddit.subreddit.value_counts()
```

```python
pd.set_option('display.max_colwidth', None)
reddit[['subreddit', 'body',]].sample(20)
```

```python
# Compare with white supremacist dataset
# Load white supremacist dataset to count posts over time
import datetime

path = '../tmp/white_supremacist_train_corpus.pkl'
ws_data = pd.read_pickle(path)

# Select forum data, Group by year
forum_posts = ws_data.query('domain=="forum"')
yearly = forum_posts.groupby(by=forum_posts.timestamp.dt.year)['text'].count()
lookup = pd.DataFrame(yearly)
lookup['begin'] = pd.to_datetime(yearly.index.astype(int).astype(str), format='%Y')
lookup['end'] = lookup.begin.shift(-1, fill_value = datetime.datetime(2020,1,1))
lookup.index.name = 'year'
lookup.index = lookup.index.astype(int)
lookup.rename(columns={'text': 'post_count'}, inplace=True)
lookup
```

```python
reddit_yearly = reddit.groupby(reddit.timestamp.dt.year)['body'].count()
reddit_yearly
```

## Check scraped Medium articles

```python
# Load data
import os
import re
import pandas as pd

dirpath = '../data/antiracist/medium/'
dfs = []
for fname in [f for f in os.listdir(dirpath) if f.endswith('.jsonl')]:
    tag = re.split(r'_articles\d?\.', fname)[0]
    fpath = os.path.join(dirpath, fname)
    dfs.append(pd.read_json(fpath, orient='records', lines=True).assign(tag=tag))
data = pd.concat(dfs)

# Filter out empty articles
print(len(data))
data = data[data.title != data.text]
data.info()
data.tag.value_counts()
```

```python
from IPython.display import display

pd.set_option('display.max_colwidth', None)
for tag in data.tag.unique():
    print(tag)
    display(data.query('tag==@tag').sample(5))
```

## Compare with white supremacist long-form data

```python
# See count by year
data_yearly = data.groupby(data.date.dt.year)['text'].count()
data_yearly
```

```python
# Process
import nltk

def tokenize_lowercase(inp):
    """ Tokenize and lowercase text """
    return ' '.join(nltk.word_tokenize(str(inp))).lower()
```

```python
from tqdm.notebook import tqdm

data['processed_text'] = [tokenize_lowercase(t) for t in tqdm(data['text'])]
data[['processed_text', 'text']].sample(3)
```

```python
# See how many white supremacist articles there are to match
path = '../tmp/white_supremacist_train_corpus.pkl'
ws_data = pd.read_pickle(path)
ws_data.domain.unique()
```

```python
# Select long-form data, Group by year
ws_long = ws_data.query('domain=="long-form"').copy()
ws_yearly = ws_long.groupby(by=ws_long.timestamp.dt.year)['text'].count()
ws_yearly
```

```python
# Word count/year
ws_long['word_count'] = ws_long.text.str.split().str.len()
ws_long.groupby(by=ws_long.timestamp.dt.year).word_count.sum()
```

```python
data['word_count'] = data.processed_text.str.split().str.len()

article_wc = data.groupby(data.date.dt.year).agg({'word_count': ['sum', 'mean']})

# Compare number of words between white supremacist dataset long-form data and article data
print('Means')
print(ws_long.word_count.mean())
print(data.word_count.mean()) # but BERT truncates posts to 512 tokens, so will max out at that
print('Sums')
print(ws_long.word_count.sum())
print(data.word_count.sum())

ws_wc = ws_long.groupby(ws_long.timestamp.dt.year).agg({'word_count': ['sum', 'mean']})

comparison = pd.concat([ws_wc, article_wc], axis=1, keys=['white_supremacist', 'articles'])
comparison.dropna(subset=[('articles', 'word_count', 'sum')], inplace=True)
comparison['article_percentage'] = comparison[('articles', 'word_count', 'sum')]/comparison[('white_supremacist', 'word_count', 'sum')]
comparison
```
