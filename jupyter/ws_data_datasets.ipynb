```python
import nltk
import re

def remove_special(text):
    text = text.replace('>', '')
    text = re.sub(r'\d{7,}', '', text)
    text = re.sub(r'\S+(?:\.com|\.org|\.edu)\S*|https?:\/\/\S*', '', text) # Remove URLs
    return text


def process_rieger2021(text):
    """ Preprocess Rieger+ 2021 4chan, 8chan, t_D data """
    # Remove special characters
    text = remove_special(str(text))
    # Tokenize
    text = ' '.join(nltk.word_tokenize(str(text))).lower()
    return text

data['text'] = data['Text'].map(process_rieger2021)
data[['text', 'Text']]
```

```python
# Test processing
from nltk.tokenize import TweetTokenizer
import re

tokenizer = TweetTokenizer(strip_handles=True)

def remove_urls(text):
    """ Remove URLs from a text """
    return re.sub(r'\S+(?:\.com|\.org|\.edu)\S*|https?:\/\/\S*', '', text)

def remove_nonlatin(text):
    """ Remove non-Latin characters from a text """
    return re.sub(u'[^\\x00-\\x7F\\x80-\\xFF\\u0100-\\u017F\\u0180-\\u024F\\u1E00-\\u1EFF]', '', text)

def process_tweet_text(text, tokenizer):
    """ Process tweet text (if user mentions and urls are not available) """
    new_text = text
    new_text = remove_urls(new_text)
    new_text = remove_nonlatin(new_text)
    new_text = ' '.join(tokenizer.tokenize(new_text))
    new_text = new_text.replace('RT : ', '')
    new_text = new_text.replace('_', '') # specific to Rieger+2021 dataset issues
    new_text = re.sub('http|https', '', new_text) # specific to Rieger+2021 dataset issues
    return new_text.lower()

tweets['processed'] = [process_tweet_text(text, tokenizer) for text in tweets['text']]
tweets[['processed', 'text']].sample(20)
```

# Pruden+2022 white supremacist long texts

```python
# Load data
import os
import pandas as pd
import re

dfs = []
dirpath = '/storage2/mamille3/white_supremacist_lang/data/pruden2022/'
for fname in sorted(os.listdir(dirpath)):
    print(fname)
    fpath = os.path.join(dirpath, fname)
    if fname == 'AndersBehringBreivikManifesto.txt':
        with open(fpath, encoding='latin-1') as f:
            text = [line.strip() for line in re.split(r'\n\s+', f.read()) if len(line.strip()) > 0]
        print(len(text))
    else:
        with open(fpath, encoding='latin-1') as f:
            text = [line.strip() for line in f.read().splitlines() if len(line.strip()) > 0]
        print(len(text))
    dfs.append(pd.DataFrame({'title': fname.split('.txt')[0], 'text': text}))
texts =  pd.concat(dfs)
texts
```

```python
# Look into standardizing units into paragraphs or similar across books and manifestos
from tqdm.notebook import tqdm
import nltk

texts['processed'] = [' '.join(nltk.word_tokenize(str(x))).lower() for x in tqdm(texts['text'])]

# Mean, sum number of words
pd.options.plotting.backend = 'plotly'
texts['num_words'] = texts.processed.str.split().str.len()
mean_words =  texts.groupby('title')['num_words'].mean()
display(mean_words)
mean_words.sort_values(ascending=False).plot.bar(title='mean words').show()

word_count =  texts.groupby('title')['num_words'].sum()
display(word_count)
word_count.sort_values(ascending=False).plot.bar(title='sum words')
```
